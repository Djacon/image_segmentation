## Load model

```python
model = YOLO('yolov8s-seg.pt')
```

```python
model.model
```

## Create segmentation function

```python
def overlay(image, mask, color, alpha=.5):
    colored_mask = np.expand_dims(mask, 0).repeat(3, axis=0)
    colored_mask = np.moveaxis(colored_mask, 0, -1)
    masked = np.ma.MaskedArray(image, mask=colored_mask, fill_value=color)
    image_overlay = masked.filled()

    return cv2.addWeighted(image, 1 - alpha, image_overlay, alpha, 0)
```

```python
def segment_image(image, with_mask=True):
    result = model(image, conf=.5, line_thickness=2)[0]

    image = result.orig_img
    for j, (x1, y1, x2, y2, conf, cls) in enumerate(result.boxes.data):
        cls = int(cls.item())
        color = COLORS[cls]

        x1, y1, x2, y2 = int(x1.item()), int(y1.item()), int(x2.item()), int(y2.item())
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)

        class_name = f"{result.names[cls]} {conf:.2f}"

        font_size, _ = cv2.getTextSize(class_name, cv2.FONT_HERSHEY_PLAIN, 1, 1)

        if y1 < 31:
            y1 += 30

        cv2.rectangle(image, (x1, y1 - 30), (x1 + font_size[0], y1), color, -1)
        cv2.putText(image, class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_PLAIN, 1, (255, 255, 255), 1, cv2.LINE_AA)

        if with_mask:
            mask = result.masks.data[j].cpu().numpy()
            mask = scale_image(mask, image.shape)
            image = overlay(image, mask, color)

    return image
```

## Testing model with our camera (skip if you dont have camera)

```python
cap = cv2.VideoCapture(0)
ret, frame = cap.read()

while ret:
    ret, frame = cap.read()

    result = segment_image(frame)
    cv2.imshow('YOLO', result)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    ret, frame = cap.read()

cap.release()
cv2.destroyAllWindows()
```
